```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
# Import essential models and functions from sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

# Q1a &b)

```python
csv_data = pd.read_csv('train.csv', header = 0 ) 
CentralAir = pd.DataFrame(csv_data['CentralAir']) 
SalePrice  = pd.DataFrame(csv_data['SalePrice'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(SalePrice, CentralAir, test_size = 0.25)
# Decision Tree using Train Data
dectree_depth3 = DecisionTreeClassifier(max_depth = 3)  # create the decision tree object
dectree_depth3.fit(X_train, y_train)                    # train the decision tree model

# Predict Legendary values corresponding to Total
y_train_pred = dectree_depth3.predict(X_train)
y_test_pred = dectree_depth3.predict(X_test)

# Plot the trained Decision Tree
from sklearn.tree import plot_tree

f = plt.figure(figsize=(12,12))
plot_tree(dectree_depth3, filled=True, rounded=True, 
          feature_names=["SalePrice"], 
          class_names=["No","Yes"])

# Plot the Confusion Matrix for Train and Test
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(confusion_matrix(y_train, y_train_pred),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(y_test, y_test_pred), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])


```

```python
from sklearn.metrics import accuracy_score, confusion_matrix



# For Tree 1
accuracy_train = accuracy_score(y_train, y_train_pred )
conf_matrix_train = confusion_matrix(y_train, y_train_pred )

# True Positives, False Positives, True Negatives, False Negatives
TP_train = conf_matrix_train[1][1]
FP_train = conf_matrix_train[0][1]
TN_train = conf_matrix_train[0][0]
FN_train = conf_matrix_train[1][0]

TPR_train = TP_train / (TP_train + FN_train)
FPR_train = FP_train / (FP_train + TN_train)

print("Train data")
print("Classification Accuracy:", accuracy_train)
print("True Positive Rate:", TPR_train)
print("False Positive Rate:", FPR_train)
print()

from sklearn.metrics import accuracy_score, confusion_matrix

# For Tree 1
accuracy_test = accuracy_score(y_test, y_test_pred )
conf_matrix_test = confusion_matrix(y_test, y_test_pred )

# True Positives, False Positives, True Negatives, False Negatives
TP_test = conf_matrix_test[1][1]
FP_test = conf_matrix_test[0][1]
TN_test= conf_matrix_test[0][0]
FN_test = conf_matrix_test[1][0]

TPR_test = TP_test/ (TP_test + FN_test)
FPR_test= FP_test / (FP_test + TN_test)

print("Test data")
print("Classification Accuracy:", accuracy_test)
print("True Positive Rate:", TPR_test)
print("False Positive Rate:", FPR_test)
print()
```

```python


dectree_depth4 = DecisionTreeClassifier(max_depth = 4)  # create the decision tree object
dectree_depth4.fit(X_train, y_train)                    # train the decision tree model
y_train_pred_depth4 = dectree_depth4.predict(X_train)
y_test_pred_depth4 = dectree_depth4.predict(X_test)

# Plot the trained Decision Tree
from sklearn.tree import plot_tree

f = plt.figure(figsize=(12,12))
plot_tree(dectree_depth4, filled=True, rounded=True, 
          feature_names=["SalePrice"], 
          class_names=["No","Yes"])

# Plot the Confusion Matrix for Train and Test
f, axes = plt.subplots(1, 2, figsize=(12, 4))
sb.heatmap(confusion_matrix(y_train, y_train_pred_depth4),
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[0])
sb.heatmap(confusion_matrix(y_test, y_test_pred_depth4), 
           annot = True, fmt=".0f", annot_kws={"size": 18}, ax = axes[1])

accuracy_train = accuracy_score(y_train, y_train_pred_depth4 )
conf_matrix_train = confusion_matrix(y_train, y_train_pred_depth4 )

```

```python
from sklearn.metrics import accuracy_score, confusion_matrix



# True Positives, False Positives, True Negatives, False Negatives
TP_train = conf_matrix_train[1][1]
FP_train = conf_matrix_train[0][1]
TN_train = conf_matrix_train[0][0]
FN_train = conf_matrix_train[1][0]

TPR_train = TP_train / (TP_train + FN_train)
FPR_train = FP_train / (FP_train + TN_train)

print("Train data")
print("Classification Accuracy:", accuracy_train)
print("True Positive Rate:", TPR_train)
print("False Positive Rate:", FPR_train)
print()

from sklearn.metrics import accuracy_score, confusion_matrix


accuracy_test = accuracy_score(y_test, y_test_pred_depth4 )
conf_matrix_test = confusion_matrix(y_test, y_test_pred_depth4 )


TP_test = conf_matrix_test[1][1]
FP_test = conf_matrix_test[0][1]
TN_test= conf_matrix_test[0][0]
FN_test = conf_matrix_test[1][0]

TPR_test = TP_test/ (TP_test + FN_test)
FPR_test= FP_test / (FP_test + TN_test)

print("Test data")
print("Classification Accuracy:", accuracy_test)
print("True Positive Rate:", TPR_test)
print("False Positive Rate:", FPR_test)
print()
```

# Depth 3 is better because its classification accuracy for test data is higher than the classification accuracy for test data of depth 4. Its true postive rate in test data is also higher than that of the true positive rate in test data for depth 4.
